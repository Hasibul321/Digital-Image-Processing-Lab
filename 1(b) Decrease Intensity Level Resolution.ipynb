### Importing the Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Function for Decreasing Intensity Level Resolution by 1-Bit

```python
def decrease_intensity_resolution(image, bit):
    level = 2**bit
    step = 255 / (level - 1)
    decreased_image = []
    
    for c in range (0, image.shape[0], 2):
        temporary_row = []
        for r in range (0, image.shape[1], 2):
            pixel = round(image[c, r] / step) * step
            temporary_row.append(pixel)
        decreased_image.append(temporary_row)
        
    return np.uint8(decreased_image)
```

### Loading the Original Image

```python
original_image = cv2.imread("./Joker.jpg", 0)
original_image = cv2.resize(original_image, (512, 512))
```

### Decreasing Intensity Level Resolution by 1-Bit up to it's Binary Format

```python
subplot_number = 8
plt.figure(figsize = (13, 7))

for bit in range(1, 9):
    decreased_image = decrease_intensity_resolution(original_image, bit)
    plt.subplot(2, 4, subplot_number)
    plt.imshow(cv2.cvtColor(decreased_image, cv2.COLOR_BGR2RGB))
    plt.title(f"{bit}-Bit Image")
    subplot_number -= 1

plt.show()
```
