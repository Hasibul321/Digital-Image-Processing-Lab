**Task 1(a): Take grayscale image of size 512x512, decrease it's spatial resolution by half every time & observe it's change when displaying in the same window size**

### Importing the Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Function for Decreasing Spatial Resolution by Half

```python
def decrease_spatial_resolution_by_half(image):
    decreased_image = []
    height, width = image.shape

    for c in range(0, height, 2):
        temporary_row = []
        for r in range(0, width, 2):
            pixel = image[c, r]
            temporary_row.append(pixel)
        decreased_image.append(temporary_row)

    return np.uint8(decreased_image)
```

### Loading the Original Image

```python
original_image = cv2.imread("./Joker.jpg", 0)
original_image = cv2.resize(original_image, (512, 512))
```

### Decreasing the Spatial Resolution by Half

```python
decreased_image = original_image.copy()
plt.figure(figsize = (13, 6))

for img_count in range(1, 9):
    plt.subplot(2, 4, img_count)
    plt.imshow(cv2.cvtColor(decreased_image, cv2.COLOR_BGR2RGB))
    height, width = decreased_image.shape
    plt.title(f"{height}x{width}")
    decreased_image = decrease_spatial_resolution_by_half(decreased_image)
    
plt.show()
```

### Decreasing the Spatial Resolution by Half with Book Example

```python
rose_image = cv2.imread("./Fig0219 Rose 1024x1024.tif", 0)
decreased_image = rose_image.copy()
plt.figure(figsize = (15, 10))

for img_count in range(1, 7):
    plt.subplot(2, 3, img_count)
    plt.imshow(cv2.cvtColor(decreased_image, cv2.COLOR_BGR2RGB))
    height, width = decreased_image.shape
    plt.title(f"{height}x{width}")
    decreased_image = decrease_spatial_resolution_by_half(decreased_image)
    
plt.show()
```
