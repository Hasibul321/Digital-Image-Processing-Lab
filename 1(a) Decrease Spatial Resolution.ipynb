### Importing the Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Image Plot Function

```python
def plot_image(image, message, subplot_no):
    plt.figure(figsize = (13, 7))
    plt.subplot(2, 4, subplot_no)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(message)
```

### Resolution Decrease Function

```python
def decrease_resolution_by_half(image):
    height, width = image.shape
    decreased_image = []
    for c in range(0, height, 2):
        pixels = []
        for r in range(0, width, 2):
            pixel = image[c, r] * 0.25 + image[c + 1, r] * 0.25 + image[c, r + 1] * 0.25 + image[c + 1, r + 1] * 0.25
            pixels.append(pixel)
        decreased_image.append(pixels)
    return np.uint8(decreased_image)
```

### Loading the Image

```python
original_image = cv2.resize(cv2.imread('Joker.jpg', 0), (512, 512))
plot_image(original_image, "Original Image", 1)
```

### Decreasing resolution by half  

```python
image = original_image
for k in range(2, 9):
    decreased_image = decrease_resolution_by_half(image)
    plot_image(decreased_image, f"{decreased_image.shape[0]}x{decreased_image.shape[1]}", k)
    image = decreased_image

plt.show()
```
