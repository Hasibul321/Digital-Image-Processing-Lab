### Importing the Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Spatial Resolution Decrease Function

```python
def decrease_spatial_resolution(image):
    height, width = image.shape
    decreased_image = []
    for c in range (0, height, 2):
        decreased_row = []
        for r in range (0, width, 2):
            pixel = (.25 * image[c, r]) + (.25 * image[c + 1, r])
            pixel += (.25 * image[c, r + 1]) + (.25 * image[c + 1, r + 1])
            decreased_row.append(pixel)
        decreased_image.append(decreased_row)
    return np.uint8(decreased_image)
```

### Loading the Original Image

```python
original_image = cv2.imread("Joker.jpg", 0)
original_image = cv2.resize(original_image, (512, 512))
```

### Decreasing the Spatial Resolution

```python
image = original_image.copy()
plt.figure(figsize = (13, 7))
for k in range(1, 9):
    plt.subplot(2, 4, k)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(f"{image.shape[0]}x{image.shape[1]}")
    image = decrease_spatial_resolution(image)
plt.show()
```
